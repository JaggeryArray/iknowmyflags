```python
# read flag list from flag_list_json/sovereign_plus_few_flags.json
import json

with open('flag_list_json/sovereign_plus_few_flags.json', 'r') as f:
    flag_list = json.load(f)

flag_list
```

```python
from io import BytesIO
from pathlib import Path

import cairosvg
import numpy as np
from PIL import Image


def svg_to_ascii_colour(svg_path, width=60):

    png = cairosvg.svg2png(url=str(svg_path))
    img = Image.open(BytesIO(png)).convert("RGBA")

    aspect = img.height / img.width
    height = max(1, int(width * aspect * 0.5))

    img = img.resize((width, height))

    out = []

    for y in range(img.height):

        line = ""

        for x in range(img.width):

            r, g, b, a = img.getpixel((x, y))

            # transparent pixel
            if a == 0:
                line += " "
                continue

            char = "█"

            line += (
                f"\033[38;2;{r};{g};{b}m"
                f"{char}"
                "\033[0m"
            )

        out.append(line)

    return "\n".join(out)
```

```python
import random

code = random.choice(list(flag_list.keys()))
country = flag_list[code]

print(svg_to_ascii_colour(Path("4x3_flags") / f"{code}.svg", width = 60))

print(f"Answer: {country}")
```
